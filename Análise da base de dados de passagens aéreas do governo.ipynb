```python
import pandas as pd
```

```python
# Criar DataFrame a partir do arquivo .xlsx (skiprows - pular 3 primeiras linhas, skipfooter - pular 3 últimas linhas)
df = pd.read_excel('201710_EmissaoPassagens_SCDP.xlsx', skiprows=3, skipfooter=3)

# 5 primeiros registros
df.head()
```

```python
# 5 últimos registros
df.tail()
```

```python
# 5 registros aleatórios
df.sample(5)
```

```python
# Dimensão do DataFrame
df.shape
```

```python
# Tipos das colunas
df.dtypes
```

```python
# Tipos e quantidades de registros por coluna
df.info()
```

```python
# Mudar tipo da coluna Data Embarque para datetime
df['Data Embarque'] = pd.to_datetime(df['Data Embarque'])
df['Data Embarque'].head()
```

```python
# Mudar tipo da coluna Data Emissão Bilhete para datetime
df['Data Emissão Bilhete'] = pd.to_datetime(df['Data Emissão Bilhete'])
df['Data Emissão Bilhete'].head()
```

```python
# Renomear coluna Nome do órgão superior para retirar os espaços
df.rename(columns={'Nome do órgão superior              ': 'Nome do órgão superior'}, inplace=True)
```

```python
# 5 maiores valores de bilhete
df.nlargest(5, 'Valor Bilhete')
```

```python
# Retirar espaços dos valores da coluna
df['Cancelado'] = df['Cancelado'].str.strip()
```

```python
# DataFrame apenas com voos cancelados
df_cancelado = df[df['Cancelado'] == 'Sim']
```

```python
# 3 maiores valores de bilhete de voos que foram cancelados
df_cancelado.nlargest(3, 'Valor Bilhete')
```

```python
# Quantidade de voos cancelados por data de emissão do bilhete
df_cancelado['Data Emissão Bilhete'].dt.year.value_counts()
```

```python
# Quantidade de voos cancelados por data de embarque
df_cancelado['Data Embarque'].dt.year.value_counts()
```

```python
# Retirar os espaços das colunas 'object'
df['Nome do órgão superior'] = df['Nome do órgão superior'].str.strip()
df['Nome do órgão solicitante da viagem'] = df['Nome do órgão solicitante da viagem'].str.strip()
df['N. PCDP'] = df['N. PCDP'].str.strip()
df['N. Reserva/Localizador'] = df['N. Reserva/Localizador'].str.strip()
df['Companhia Aérea'] = df['Companhia Aérea'].str.strip()
df['Classe Tarifária Bilhete'] = df['Classe Tarifária Bilhete'].str.strip()
df['Regra Tarifária'] = df['Regra Tarifária'].str.strip()
df['No Show'] = df['No Show'].str.strip()
df['Remarcado'] = df['Remarcado'].str.strip()
df['Situação Final Bilhete'] = df['Situação Final Bilhete'].str.strip()
```

```python
# Quantidade de voos por companhia aérea
df['Companhia Aérea'].value_counts()
```

```python
%matplotlib inline
```

```python
# Gráfico dos voos por companhia aérea
df['Companhia Aérea'].value_counts().plot.bar()
```

```python
# Quantidade de voos cancelados por companhia
df_cancelado['Companhia Aérea'].value_counts()
```

```python
# Gráficos do número de voos cancelados por companhia aérea
df_cancelado['Companhia Aérea'].value_counts().plot.bar()
```

```python
# Soma do valor gasto com bilhetes por órgão superior 
df_orgao_valor = df.groupby('Nome do órgão superior')['Valor Bilhete'].agg(['sum', 'count'])
```

```python
# Os 5 órgãos superiores com maior valor gasto em bilhetes
df_orgao_valor.nlargest(5, 'sum')
```

```python
# Os 5 órgãos superiores com menor valor gasto em bilhetes
df_orgao_valor.nsmallest(5, 'sum')
```

```python
# Soma do valor gasto com bilhetes por órgão solicitante 
df_orgao_solicitante_valor = df.groupby('Nome do órgão solicitante da viagem')['Valor Bilhete'].agg(['sum', 'count'])
```

```python
# Os 5 órgãos solicitantes com maior valor gasto em bilhetes
df_orgao_solicitante_valor.nlargest(5, 'sum')
```

```python
# Os 5 órgãos solicitantes com menor valor gasto em bilhetes
df_orgao_solicitante_valor.nsmallest(5, 'sum')
```

```python
# Gráfico com os 5 órgãos solicitantes com menor valor gasto em bilhetes
df_orgao_solicitante_valor.nsmallest(5, 'sum').plot.barh()
```

```python
# Apenas registros de No Show
df_noshow = df[df['No Show'] == 'Sim']
```

```python
# Registros de No Show por Órgão solicitante da viagem
df_noshow_orgaosol = df_noshow.groupby('Nome do órgão solicitante da viagem')['No Show'].agg(['count'])
```

```python
# Os 5 órgãos solicitantes com maior número de registros de No Show
df_noshow_orgaosol.nlargest(5, 'count')
```

```python
# Gráfico com os 5 órgãos solicitantes com maior número de registros de No Show
df_noshow_orgaosol.nlargest(5, 'count').plot.barh()
```
